# <u>Airbnb Market Analysis</U>

# Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load the Data

```python
market_analysis_path = 'market_analysis_2019.csv'
df = pd.read_csv(market_analysis_path, delimiter=';', decimal=',')
```

# Explaore the Data

```python
print("First 5 rows:")
df.head()
```

```python
print("\nDataset Info:")
df.info()
```

```python
print("\nDescriptive Statistics:")
df.describe(include='all').T
```

# Data Cleaning

```python
# Check missing values
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
#Drop rows with missing values (can also use fillna() instead)
df.dropna(inplace=True)
```

```python
# Convert data types where necessary
df["bedrooms"] = df["bedrooms"].astype(str)
df["bathrooms"] = pd.to_numeric(df["bathrooms"], errors="coerce")
df["guests"] = df["guests"].str.replace('+','', regex=False)
df["guests"] = pd.to_numeric(df["guests"], errors="coerce")
df["revenue"] = pd.to_numeric(df["revenue"], errors="coerce")
df["nightly rate"] = pd.to_numeric(df["nightly rate"], errors="coerce")
```

```python
# Convert month column to datetime
df["month"] = pd.to_datetime(df["month"], format="%Y-%m", errors='coerce')
```

```python
# Drop any remaining missing values after conversions
df.dropna(inplace=True)
```

# Data Analysis

```python
# Value counts for guests
print("\nGuests value counts:")
print(df["guests"].value_counts())
```

```python
# Group by city to get average revenue
avg_revenue_city = df.groupby("city")["revenue"].mean().sort_values(ascending=False)
print("\nAverage Revenue by City:")
print(avg_revenue_city)
```

```python
# Average nightly rate per city
avg_rate_city = df.groupby("city")["nightly rate"].mean().sort_values(ascending=False)
print("\nAverage Nightly Rate by City:")
print(avg_rate_city)
```

```python
# Guest capacity distribution
print("\nGuest Capacity Distribution:")
print(df["guests"].value_counts().sort_index())
```

```python
# Correlation matrix
print("\nCorrelation Matrix:")
print(df.corr(numeric_only=True))
```

# Data Visualization

```python
# a. Guest Capacity Bar Chart
plt.figure(figsize=(8, 5))
df["guests"].value_counts().sort_index().plot(kind="bar", color='teal')
plt.title("Listings by Guest Capacity")
plt.xlabel("Guest Capacity")
plt.ylabel("Count")
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# b. Revenue Distribution
plt.figure(figsize=(8, 5))
sns.histplot(df["revenue"], bins=30, kde=True, color="coral")
plt.title("Revenue Distribution")
plt.xlabel("Revenue ($)")
plt.ylabel("Frequency")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# c. City-wise Average Revenue
plt.figure(figsize=(10, 6))
avg_revenue_city.plot(kind="bar", color="steelblue")
plt.title("Average Revenue by City")
plt.xlabel("City")
plt.ylabel("Average Revenue ($)")
plt.grid(axis='y')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# d. Scatter: Nightly Rate vs Revenue
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x="nightly rate", y="revenue", hue="city")
plt.title("Revenue vs Nightly Rate")
plt.xlabel("Nightly Rate ($)")
plt.ylabel("Revenue ($)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# e. Heatmap: Correlation
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()
```

```python
# f. Revenue over Time
plt.figure(figsize=(10, 6))
df.groupby("month")["revenue"].sum().plot(marker='o')
plt.title("Total Revenue Over Time")
plt.xlabel("Month")
plt.ylabel("Total Revenue ($)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# g. Occupancy by Host Type
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x="host_type", y="occupancy")
plt.title("Occupancy Distribution by Host Type")
plt.xlabel("Host Type")
plt.ylabel("Occupancy Rate")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Draw Conclusions

```python
print("\n=== Key Insights ===")
print(f"- City with highest average revenue: {avg_revenue_city.idxmax()} (${avg_revenue_city.max():.2f})")
print(f"- City with highest average nightly rate: {avg_rate_city.idxmax()} (${avg_rate_city.max():.2f})")
print(f"- Most common guest capacity: {df['guests'].mode()[0]}")
```
